```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x=[1.2,2.5,3.2,4.5,5.2,6.5,7.2,8.3,9.6,4.5,6.2,7.9,8.6,6.8,4.1,7.8,8.9,5.4]
y=[125,250,330,465,562,654,789,859,958,452,635,741,852,654,412,753,851,589]
```

```python
len(y)
```

```python
X=np.array(x)
Y=np.array(y)
```

```python
m=10
c=5
lr= 0.02
for i in range(200):
    yhat=X*m+c
    E=np.mean(np.square(Y-yhat))
    m=m- lr*2*np.mean((yhat-Y)*X)
    c=c-lr*2*np.mean(yhat-Y)
    print(E)
```

```python
print(m)
print(c)
```

```python
sns.scatterplot(X,Y)
```

```python
def predict(a):
    ypred=a*m+c
    return ypred
```

```python
yp=predict(X)
```

```python
plt.scatter(X,Y)
plt.plot(X,yp,c='r')
plt.show()
```
